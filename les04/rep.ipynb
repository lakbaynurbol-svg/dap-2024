```python
import numpy as np
import pandas as pd
import re

# 1. Создание DataFrame
df = pd.read_csv('titanic.csv', index_col='PassengerId')

# 2. Первые 6 строк
print(df.head(6))

# 3. Описание данных
print(df.describe())
print(df.info())

# 4. Количество мужчин и женщин
print(df['Sex'].value_counts())

# 5. Распределение Pclass и по полу
print(df['Pclass'].value_counts())
print(df.groupby(['Sex','Pclass']).size().unstack(fill_value=0))
print("Мужчин 2-го класса:", df.groupby(['Sex','Pclass']).size().loc['male',2])

# 6. Медиана и стандартное отклонение Fare
print("Медиана:", round(df['Fare'].median(),2))
print("Ст. отклонение:", round(df['Fare'].std(),2))

# 7. Выживаемость <30 и >60
print("Доля выживших <30:", df[df['Age']<30]['Survived'].mean())
print("Доля выживших >60:", df[df['Age']>60]['Survived'].mean())

# 8. Выживаемость мужчин и женщин
print(df.groupby('Sex')['Survived'].mean())

# 9. Самое популярное мужское имя
def extract_male_firstname(full):
    m = re.search(r'Mr\.?\s+([A-Za-z\-]+)', full)
    return m.group(1) if m else None
names = df[df['Sex']=='male']['Name'].apply(extract_male_firstname).dropna()
print
```

```python
###
#Дополнительные вопросы
1. Медиана — значение, делящее ряд пополам.
Пример: np.median([1,3,5,7,9]) = 5

2. Дисперсия — средний квадрат отклонений от среднего.
Пример: np.var([1,3,5,7,9], ddof=1) = 10

3. Среднеквадратичное отклонение — корень из дисперсии.
Пример: np.std([1,3,5,7,9], ddof=1) ≈ 3.16

4. Series — одномерная структура (индекс + значения).
DataFrame — двумерная таблица (несколько Series).

5. Индекс в Series можно изменить после создания.

6. Сложение Series:
- одинаковые индексы → поэлементное сложение
- разные индексы → выравнивание, NaN где нет совпадений
- разные типы → приведение типов

7. DataFrame — таблица, Series — один столбец.

8. Чтобы по умолчанию показывалось 8 строк:
pd.set_option('display.max_rows', 8)

9. iloc — индексация по позиции.
loc — индексация по меткам.

10. В DataFrame можно добавить Series как новый столбец.

11. Для CSV и JSON доп. библиотеки не нужны.
Для Excel нужны (openpyxl, xlrd).
###
```
