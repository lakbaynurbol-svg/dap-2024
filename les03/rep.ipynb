```python
### 1. Массив 10x10 со случайными значениями, найти минимум и максимум
import numpy as np

z = np.random.random((10,10))
print("Минимум:", z.min())
print("Максимум:", z.max())
```

```python
### 2. 8x8 матрица в шахматном порядке
z = np.zeros((8,8), dtype=int)
z[1::2, ::2] = 1
z[::2, 1::2] = 1
print(z)
```

```python
### 3. Поменять знак у элементов между 3 и 8
z = np.arange(11)
z[(z > 3) & (z < 8)] *= -1
print(z)
```

```python
### 4. Вычесть среднее значение строки из элементов строки
z = np.random.rand(5, 10)
z = z - z.mean(axis=1, keepdims=True)
print(z)
```

```python
### 6. Преобразовать точки из декартовой системы в полярную
z = np.arange(25).reshape(5,5)
z[[0,1]] = z[[1,0]]
print(z)
```

```python
### 7. Заменить максимальный элемент массива на ноль
z = np.random.random(10)
z[z.argmax()] = 0
print(z)
```

```python
### 8. Добавить 1 к элементам с индексами из другого массива
z = np.ones(10)
i = np.random.randint(0, len(z), 20)
z[np.unique(i)] += 1
print(z)
```

```python
### 9. Новый вектор с тремя нулями между значениями
z = np.array([1,2,3,4,5])
new_z = np.zeros(len(z)*4 - 3, dtype=int)
new_z[::4] = z
print(new_z)
```

```python
### 10. Найти ближайшее число к заданному
z = np.arange(100)
v = np.random.uniform(0,100)
closest = z[np.abs(z-v).argmin()]
print("v =", v, "Ближайшее:", closest)
```
